```python
import duckdb
import pandas as pd

# Koppling mot nuvarande databas
conn = duckdb.connect("music_warehouse.duckdb")

# 1) Hämta alla tabellnamn
tables_df = conn.execute("SHOW TABLES").df()
tables = tables_df["name"].tolist()
print(f"Hittade tabeller: {tables}\n")

for table in tables:
    # 2)) Hämta schema (vilka kolumner som finns och deras datatyper)
    print(f"--- SCHEMA FÖR TABELL: {table} ---")
    schema_df = conn.execute(f"DESCRIBE {table}").df()
    # printa bara kolumnnamn och datatyp för att hålla det rent
    print(schema_df[["column_name", "column_type"]].to_string(index=False))

    # 3) Hämta 3 rader data för att se hur innehållet faktiskt ser ut
    print(f"\n--- SAMPLE DATA ({table}) ---")
    sample_df = conn.execute(f"SELECT * FROM {table} LIMIT 3").df()
    print(sample_df.to_string(index=False))
    print("\n" + "=" * 50 + "\n")
```
