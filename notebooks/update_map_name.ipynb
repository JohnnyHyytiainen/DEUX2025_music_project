## Country-analysis
- historical_charts has countries
- spotify_daily has countries
- top_200 has countries
- dim_geography is a mapping-table with countries and country-codes.

### Action: Clean countrynames to be able to map them to dim_geography

```python
df_historical = pd.read_sql_query("SELECT * FROM historical_charts", conn)

unique_countries = df_historical['country'].unique()

print("Unique country-value in historical_charts:")
print(unique_countries)
```

### Function for cleaning name

```python
def get_iso_code(name):
    if pd.isna(name) or str(name).strip().lower() == 'unknown': 
        return 'Unknown'
    if str(name).strip().title() == 'Global': 
        return 'Global'
    try:
        # fuzzy match hittar "Suecia" och plockar ut .alpha_2 som är "SE"
        return pycountry.countries.search_fuzzy(name)[0].alpha_2
    except:
        # Om den absolut inte hittar något, returnera Unknown
        return 'Unknown'
```

### Update country

```python
# Create a mapping dictionary
mapping_dict = {}
for country in unique_countries:
    mapping_dict[country] = get_iso_code(country)

print("Mapping dictionary (original -> standardized):")
print(mapping_dict)

# Apply the mapping to the entire dataframe
df_historical['iso_code'] = df_historical['country'].map(mapping_dict)

# Save to DB
df_historical.to_sql('historical_charts_cleaned', conn, if_exists='replace', index=False)

print(f"Country names standardized and saved to '{df_historical.shape[0]}' rows in the 'historical_charts_cleaned' table in the database.")
```

```python
df_top200 = pd.read_sql_query("SELECT * FROM top_200", conn)

unique_countries_top200 = df_top200['nationality'].unique()

print("Unique country-value in top200:")
print(unique_countries_top200)
```

```python
print("Översätter namn till ISO-koder...")
mapping_dict = {}
for name in unique_countries_top200:
    mapping_dict[name] = get_iso_code(name)

# 5. Skapa den nya kolumnen i dataframe:n genom att mappa de gamla namnen
df_top200['iso_code'] = df_top200['nationality'].map(mapping_dict)

# 6. (Data Engineer Mode) Städa bort överflödig textdata för att spara plats
# Nu när vi har iso_code behöver vi inte original-namnet eller kontinenten här!
columns_to_drop = ['nationality', 'continent'] 

# Kollar så kolumnerna faktiskt finns innan vi droppar dem
existing_cols_to_drop = [col for col in columns_to_drop if col in df_top200.columns]
df_top200 = df_top200.drop(columns=existing_cols_to_drop)

# 7. Spara den tvättade tabellen
table_name = 'top_200_cleaned'
df_top200.to_sql(table_name, conn, if_exists='replace', index=False)

print(f"✅ Klar! Tabellen '{table_name}' är nu städad, har ISO-koder och är redo för Star Schemat.")
```

```python
def get_continent(iso_code):
    # Hantera vår specialare för Spotify
    if pd.isna(iso_code) or str(iso_code).strip().lower() == 'unknown':
        return 'Unknown'
    if str(iso_code).strip().title() == 'Global':
        return 'Worldwide'
    
    try:
        # Steg 1: Få kontinent-koden (t.ex. 'SE' -> 'EU')
        continent_code = pc.country_alpha2_to_continent_code(iso_code.upper())
        # Steg 2: Översätt koden till fullständigt namn (t.ex. 'EU' -> 'Europe')
        continent_name = pc.convert_continent_code_to_continent_name(continent_code)
        return continent_name
    except:
        # Om koden inte finns i databasen (t.ex. konstiga territorier)
        return 'Other'
```
