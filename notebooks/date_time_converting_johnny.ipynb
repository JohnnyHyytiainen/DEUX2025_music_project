```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import duckdb

# Mina konstanter rörs ej (Skrivs med ALL CAPS i Python)
FILE_PATH = "../data/raw/spotify_charts.parquet"

df_team = pd.read_parquet(FILE_PATH)
```

```python
# ==========================================
# 1. KONVERTERA DATUM (Str -> Datetime)
# ==========================================
# errors='coerce' är ett bra. Om det finns skräpvärden typ "okänt datum"
# gör den om det till NaT (Not a Time) istället för att krascha hela koden.
df_team['date'] = pd.to_datetime(df_team['date'], errors='coerce')

# ==========================================
# 2. VALIDERINGS-CHECK
# ==========================================
print("=== NYA DATATYPER ===")
print(df_team.dtypes)

print("\n=== FINNS DET NÅGRA NULL VÄRDEN ===")
print(df_team.isnull().sum())
```
